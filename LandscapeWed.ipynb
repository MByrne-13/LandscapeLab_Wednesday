```r
dat <- read.csv("Wednesday.csv")

#Model Aggregation as function of % forest cover
dat$FC2 <- dat$ForestCover^2
#LM
AL <- lm(A~ForestCover, data=dat)
#PM
AP <- lm(A~ForestCover + FC2, data=dat)

#data.frame of R-square results

Mod_Results <- data.frame(Model = c("linear", "polynomial"), 
                          R_square = c(summary(AL)$adj.r.squared, summary(AP)$adj.r.squared))

print("Model Fit Results: Aggregation ~ Forest")
print("")
print(Mod_Results)
```

```r
MT <- readline(prompt = "Which model fit best, linear or polynomial: ")

if (MT == "linear") {
  newdat <- data.frame(ForestCover = seq(from=0, to=100, by=1))
  
  Apred <- predict(AL, newdata = newdat)
  Apred<- data.frame(FC = newdat$ForestCover, AP = Apred)
  
  plot(A~ForestCover, data = dat, pch=19, xlim=c(0,100), ylab = "Aggregation", xlab="% Forest",
       main="Effect of % Forest on Aggregation")
  lines(AP~FC, data=Apred, lwd=2, col="red")
} else if (MT == "polynomial") {
  newdat <- data.frame(ForestCover = seq(from=0, to=100, by=1), FC2 = seq(from=0, to=100, by=1)^2)
  
  Apred <- predict(AP, newdata = newdat)
  Apred<- data.frame(FC = newdat$ForestCover, AP = Apred)
  
  plot(A~ForestCover, data = dat, pch=19, xlim=c(0,100), ylab = "Aggregation", xlab="% Forest",
       main="Effect of % Forest on Aggregation")
  lines(AP~FC, data=Apred, lwd=2, col="red")
}
```

```r
####################################################################
#Model Edge as function of % forest cover
#LM
EL <- lm(E~ForestCover, data=dat)
#PM
EP <- lm(E~ForestCover + FC2, data=dat)

#data.frame of R-square results

Mod_ResultsE <- data.frame(Model = c("linear", "polynomial"), 
                         R_square = c(summary(EL)$adj.r.squared, summary(EP)$adj.r.squared))

print("Model Fit Results: Edge ~ Forest")
print("")
print(Mod_ResultsE)
```

```r
####
MTE <- readline(prompt = "Which model fit best, linear or polynomial: ")

if (MTE == "linear") {
  newdat <- data.frame(ForestCover = seq(from=0, to=100, by=1))
  
  Epred <- predict(EL, newdata = newdat)
  Epred<- data.frame(FC = newdat$ForestCover, P = Epred)
  
  plot(E~ForestCover, data = dat, pch=19, xlim=c(0,100), ylab = "Edge", xlab="% Forest",
       main="Effect of % Forest on Edge")
  lines(P~FC, data=Epred, lwd=2, col="red")
} else if (MTE == "polynomial") {
  newdat <- data.frame(ForestCover = seq(from=0, to=100, by=1), FC2 = seq(from=0, to=100, by=1)^2)
  
  Epred <- predict(EP, newdata = newdat)
  Epred<- data.frame(FC = newdat$ForestCover, P = Epred)
  
  plot(E~ForestCover, data = dat, pch=19, xlim=c(0,100), ylab = "Edge", xlab="% Forest",
       main="Effect of % Forest on Edge")
  lines(P~FC, data=Epred, lwd=2, col="red")
}
```
